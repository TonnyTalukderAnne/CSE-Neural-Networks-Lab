```python
# 0432220005101050
# Tonny Talukder
# ----------------------------------
# Single Layer Perceptron
# Binary Classification: Student Pass / Fail
# ----------------------------------


# ---------- Task 1: Dataset Creation ----------
# Each data point = [study_hours, attendance_percentage]


data = [
    [2, 50],
    [3, 60],
    [4, 65],
    [5, 70],
    [6, 75],
    [7, 80],
    [1, 40],
    [8, 90],
    [6, 68],
    [9, 85],
    [4, 72],
    [5, 65],
    [10, 95],
    [2, 55],
    [7, 78]
]

labels = []

for item in data:
    if item[0] >= 5 and item[1] >= 70:
        labels.append(1)   # Pass
    else:
        labels.append(0)   # Fail
```

```python
# ---------- Task 2: Perceptron Initialization ----------
w1 = 0.1
w2 = 0.1
bias = 0.1
learning_rate = 0.01
```

```python
# ---------- Task 3: Activation Function ----------
def step_function(value):
    if value >= 0:
        return 1
    else:
        return 0
```

```python
# ---------- Task 4: Training ----------
epochs = 10

for epoch in range(epochs):
    total_error = 0

    for i in range(len(data)):
        x1 = data[i][0]
        x2 = data[i][1]
        target = labels[i]

        weighted_sum = (w1 * x1) + (w2 * x2) + bias
        prediction = step_function(weighted_sum)

        error = target - prediction
        total_error += abs(error)

        # Weight update rule
        w1 = w1 + learning_rate * error * x1
        w2 = w2 + learning_rate * error * x2
        bias = bias + learning_rate * error

    print("Epoch:", epoch + 1, "Total Error:", total_error)
```

```python
# ---------- Task 5: User Input Testing ----------
print("\nStudent Result Prediction")

study_hours = float(input("Enter study hours per day: "))
attendance = float(input("Enter attendance percentage: "))

result_sum = (w1 * study_hours) + (w2 * attendance) + bias
result = step_function(result_sum)

if result == 1:
    print("Prediction: The student is likely to PASS")
else:
    print("Prediction: FAIL")
```
